## 데이터 다운로드하기

````python
import os
import tarfile
import urllib.request

#데이터 다운로드 위치
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/rickiepark/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

#데이터 다운로드 함수 정의
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
````

````python
#데이터 다운로드 실행
fetch_housing_data()
````

````python
#판다스로 데이터 읽기

import pandas as pd

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)
    
#모든 데이터를 담은 판다스 데이터프레임 객체 반환
````

## 테스트 세트 만들기

````python
# 노트북의 실행 결과가 동일하도록
# skip
np.random.seed(42)
````

````python
# 무작위 샘플링에 의한 훈련/테스트 분할
# 난수 초기값 지정
# 20%의 테스트, 80%의 훈련
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(housing_182223, test_size=0.2, random_state=42)
````

````python
test_set.head()
````

````python
# 계층적 샘플링
# 중간소득(median_income)이 중간주택가격을 예측하는데 매우중요(전문가의견)
# 중간소득의 카테고리빈도에 따라 훈련/테스트세트를 구성
# 중간소득이 1.5~6 사이에 모여있음
# 계층별로 충분한 샘플수 필요, 너무 많은 계층으로 나누면 안됨

housing_182223["median_income"].hist()
````

````python
# pd.cut으로 5개의 카테고리 생성
# 1~5까지 레이블

housing_182223["income_cat"] = pd.cut(housing_182223["median_income"],
                               bins=[0., 1.5, 3.0, 4.5, 6., np.inf],
                               labels=[1, 2, 3, 4, 5])
````

````python
#카테고리별 빈도수

housing_182223["income_cat"].value_counts()
````

````python
# x축 : 1~5 레이블
housing_182223["income_cat"].hist()
````

````python
#stratifiedshfflesplit

from sklearn.model_selection import StratifiedShuffleSplit

#StratifiedShuffleSplit을 이용해 소득 카테고리를 기반으로 계층적 샘플링
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing_182223, housing_182223["income_cat"]):
    strat_train_set = housing_182223.loc[train_index]
    strat_test_set = housing_182223.loc[test_index]
````

````python
# 소득카테고리의 비율

strat_test_set["income_cat"].value_counts() / len(strat_test_set)
````

````python
# 전체데이터에서 카테고리별 빈도수
#위 셀과 비율이 거의 비슷하게 나온다는 것을 확인

housing_182223["income_cat"].value_counts() / len(housing_182223)
````

````python
# skip
#계층 샘플링과 무작위 샘플링 비교

#함수 인자로 넣은 데이터에서 전체 데이터에 대한 income_cat 데이터의 비율을 구해 반환하는 함수
def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

#무작위 샘플링 실행
train_set, test_set = train_test_split(housing_182223, test_size=0.2, random_state=42)

#샘플링 결과별로 income_cat비율을 비교 출력
compare_props = pd.DataFrame({
    "Overall": income_cat_proportions(housing_182223),
    "Stratified": income_cat_proportions(strat_test_set),
    "Random": income_cat_proportions(test_set),
}).sort_index()
compare_props["Rand. %error"] = 100 * compare_props["Random"] / compare_props["Overall"] - 100
compare_props["Strat. %error"] = 100 * compare_props["Stratified"] / compare_props["Overall"] - 100
````

````python
# skip
#샘플링 결과별로 income_cat비율을 비교 출력
#계층 샘플링은 테스트셋 비율이 전체 데이터셋과 거의 같지만, 무작위 샘플링은 테스트셋 비율이 많이 달라짐
compare_props
````

````python
# income_cat 삭제 (원래 상태로 만들기)
#income_cat은 계층 샘플링을 위해 만들어 준 것이었기 때문에 샘플링 완료 후 다시 삭제

for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
````

## 데이터 정제

책에 소개된 세 개의 옵션은 다음과 같습니다:

```
housing_182223.dropna(subset=["total_bedrooms"])    # 옵션 1(해당 구역을 제거)
housing_182223.drop("total_bedrooms", axis=1)       # 옵션 2(전체 특성을 삭제)
median = housing_182223["total_bedrooms"].median()  # 옵션 3(어떤 값으로 채움, 여기서는 평균값으로)
housing_182223["total_bedrooms"].fillna(median, inplace=True)
```

각 옵션을 설명하기 위해 주택 데이터셋의 복사본을 만듭니다. 이 때 적어도 하나의 열이 비어 있는 행만 고릅니다. 이렇게 하면 각 옵션의 정확한 동작을 눈으로 쉽게 확인할 수 있습니다.

````python
# isnull()을 이용하여 null인 항목 살펴보기
# NaN : Not a Number

sample_incomplete_rows = housing_182223[housing_182223.isnull().any(axis=1)].head()
sample_incomplete_rows
````

## 나만의 변환기(파이프라인)

추가 특성을 위해 사용자 정의 변환기를 만들어보죠:

````python
# skip

from sklearn.base import BaseEstimator, TransformerMixin

# 열 인덱스
rooms_ix, bedrooms_ix, population_ix, households_ix = 3, 4, 5, 6

#앞에서 나온 조합 특성 3개를 추가하는 변환기
#나중에 파이프라인에 들어갈 예정
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True): # *args 또는 **kargs 없음
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # 아무것도 하지 않습니다
    def transform(self, X):
        rooms_per_household = X[:, rooms_ix] / X[:, households_ix]
        population_per_household = X[:, population_ix] / X[:, households_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing_182223.to_numpy())
````

책에서는 간단하게 인덱스 (3, 4, 5, 6)을 하드코딩했지만 다음처럼 동적으로 처리하는 것이 더 좋습니다:

````python
# skip

col_names = "total_rooms", "total_bedrooms", "population", "households"
rooms_ix, bedrooms_ix, population_ix, households_ix = [
    housing_182223.columns.get_loc(c) for c in col_names] # 열 인덱스 구하기
````

또한 `housing_extra_attribs`는 넘파이 배열이기 때문에 열 이름이 없습니다(안타깝지만 사이킷런을 사용할 때 생기는 문제입니다). `DataFrame`으로 복원하려면 다음과 같이 할 수 있습니다:

````python
# skip

housing_extra_attribs = pd.DataFrame(
    housing_extra_attribs,
    columns=list(housing_182223.columns)+["rooms_per_household", "population_per_household"],
    index=housing_182223.index)
housing_extra_attribs.head()
````

## 변환 파이프라인

수치형 특성을 전처리하기 위해 파이프라인을 만듭니다:

````python
# 파이프라인 : 연속된 단계를 나타냄
# 이름/추정기쌍으로 목록입력

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

#수치형 특성에 필요한 전처리를 모아서 하나의 파이프라인으로 만듦
num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")), #null에 중간값을 추가
        ('attribs_adder', CombinedAttributesAdder()), #새로운 특성 추가
        ('std_scaler', StandardScaler()), #특성 스케일링
    ])

housing_num_tr = num_pipeline.fit_transform(housing_num)
````

````python
housing_num_tr
````

````python
#하나의 변환기로 각 열마다 다른 변환 적용
#범주형 열과 수치형 열에 각각 적절헌 변환을 적용할 수 있음

from sklearn.compose import ColumnTransformer

# 수치형 열이름 리스트
num_attribs = list(housing_num)

# 범주형 열이름 리스트
cat_attribs = ["ocean_proximity"]

# ColumnTransformer 객체 리스트
# 이름, 변환기, 변환기 적용될 열 이름(인덱스)
# num_pipeline은 위에서 정의

#범주형, 수치형 모두 한꺼번에 전처리할 수 있는 파이프라인
full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs), #위에서 만든 수치형 파이프라인
        ("cat", OneHotEncoder(), cat_attribs), #OneHotEncoding(범주형 데이터 다루는 인코딩)
    ])

# num_pipeline은 밀집행렬, OneHotEncoder()은 희소행렬
# ColumnTransformer이 밀집도( 0이 아닌 원소의 비율)가 임계값보다 낮으면 희소행렬 반환
# 밀집도가 있어서 밀집행렬 반환


# 파이프라인을 housing 데이터에 적용
housing_182223_prepared = full_pipeline.fit_transform(housing_182223)
````

## 그리드 탐색(GridSearchCV)

````python
#Grid Search를 이용해 하이퍼파라미터를 조정

from sklearn.model_selection import GridSearchCV

param_grid = [
    # 12(=3×4)개의 하이퍼파라미터 조합을 시도합니다.
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    # bootstrap은 False로 하고 6(=2×3)개의 조합을 시도합니다.
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
  ]

# n_estimators : 결정트리 갯수
# max_features : feature 수

#  {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]} : 딕셔너리
# 3 일때 2,4,6,8 조합 실행

# param_grid 설정에 따라 사이킷런이 
# 첫 번째 dict에 있는 n_estimators와 max_features의 하이퍼파라미터 조합인 3X4=12개를 평가
# 두 번째 dict에 있는 하이퍼파라미터의 조합인 2X3=6개를 시도(단, bootstrap=False)
# 총 12+6 = 18번의 조합 탐색. 


# 랜덤포레스트로 그리드탐색

forest_reg = RandomForestRegressor(random_state=42)

# 다섯 개의 폴드로 훈련하면 총 (12+6)*5=90번의 훈련이 일어납니다.

grid_search = GridSearchCV(forest_reg, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
grid_search.fit(housing_182223_prepared, housing_labels)
````

 랜덤 탐색(RandomizedSearchCV)

````python
# 랜덤하게 탐색공간을 탐색
# 탐색공간이 클 경우 그리드서치보다 추천
# 설정값이 연속형일 경우 추천
# 사이파이의 randint를 이용하여 범위내의 정수값을 균등분포로 추출 
#GridSearchCV와 거의 같은 방식으로 사용하지만,
#가능한 모든 조합을 시도하는 대신 각 반복마다 하이퍼파라미터에 임의의 수를 대입하여 지정한 횟수만큼 평가

from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {
        'n_estimators': randint(low=1, high=200),
        'max_features': randint(low=1, high=8),
    }

# 탐색 반복수n_iter=10

forest_reg = RandomForestRegressor(random_state=42)
rnd_search = RandomizedSearchCV(forest_reg, param_distributions=param_distribs,
                                n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)
rnd_search.fit(housing_182223_prepared, housing_labels)
````

````python
#랜덤 탐색의 스코어(RMSE)
cvres = rnd_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
````

## 테스트 세트로 시스템 평가하기

````python
#최종 모델
final_model = grid_search.best_estimator_

#테스트 셋에서 median_house_value 삭제(정답이니까)
X_test = strat_test_set.drop("median_house_value", axis=1)
#median_house_value(정답)를 따로 저장
y_test = strat_test_set["median_house_value"].copy()

# 훈련세트가 아니므로 fit_transform 이 아니라 transform 사용
# 테스트 셋을 파이프라인을 통해 전처리
X_test_prepared = full_pipeline.transform(X_test)
# 최종 모델을 이용해 테스트 셋을 예측
final_predictions = final_model.predict(X_test_prepared)

#모델 최종 평가(RMSE)
final_mse = mean_squared_error(y_test, final_predictions)
final_rmse = np.sqrt(final_mse)

# 테스트 데이터에 적용한 값이 더 낮은 경우가 일반적
# 테스트 데이터를 위해서 다시 하이퍼 파라미터 조정은 바람직하지 않음
````

````python
#모델 최종 평가(RMSE)
final_rmse
````

테스트 RMSE에 대한 95% 신뢰 구간을 계산할 수 있습니다:

````python
from scipy import stats

#95% 신뢰구간
confidence = 0.95

# 모든 샘플에 대한 제곱오차
squared_errors = (final_predictions - y_test) ** 2

# final_predictions 넘파이배열, y_test : 시리즈(판다스데이터프레임) -> 결과는 시리즈
# squared_errors 샘플갯수
# scipy.stats.t.interval()를 사용해 일반화 오차의 95% 신뢰구간을 계산할 수 있음
np.sqrt(stats.t.interval(confidence, len(squared_errors) - 1,
                         loc=squared_errors.mean(), #평균값
                         scale=stats.sem(squared_errors))) # 표준오차 계산

# 신뢰구간
````

다음과 같이 수동으로 계산할 수도 있습니다:

````python
#95% 신뢰구간 수동계산
#위 셀과 같은 결과를 보임

m = len(squared_errors)
mean = squared_errors.mean()
tscore = stats.t.ppf((1 + confidence) / 2, df=m - 1)
# delta degree of freedom, 기본값이 다를수 있으므로 명시적으로 표시
# numpy.std(ddof=0)
tmargin = tscore * squared_errors.std(ddof=1) / np.sqrt(m) 
np.sqrt(mean - tmargin), np.sqrt(mean + tmargin)
````

또는 t-점수 대신 z-점수를 사용할 수도 있습니다:

````python
# t-점수 대신 z-점수를 사용하여 95% 신뢰구간 계산

zscore = stats.norm.ppf((1 + confidence) / 2)

# squared_errors.std(ddof=1) / np.sqrt(m) : 표준오차
zmargin = zscore * squared_errors.std(ddof=1) / np.sqrt(m)
np.sqrt(mean - zmargin), np.sqrt(mean + zmargin)
````
